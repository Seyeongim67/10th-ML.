- 다음과 같이 학생별 각 과목의 점수가 담긴 이차원 데이터가 주어졌습니다.

- 이 문제의 목표는 학생별 총점과 각 과목의 최고점을 계산하는 것입니다.

#### **1-4-1. 주관식 질문 (2점)**
- 화연이는 데이터를 이차원 리스트로 저장하기 위해 다음과 같은 파이썬 코드를 작성했습니다.
  ```
  # 점수 데이터가 담긴 이차원 리스트
  scores = [
      ["60", "85", "50"],
      ["20", "40", "30"],
      ["100", "80", "95"]
  ]
  ```

- 화연이의 코드에서는 `scores` 데이터가 **점수 계산에 적합하지 않은 형태로** 저장되어 있습니다.

❓ **그 이유는 무엇일까요? 이 문제를 해결하려면 어떻게 해야 하는지 해결 방법을 간략하게 설명해주세요.**

- **예시 답안**

주어진 데이터는 점수가 숫자가 아닌 `문자열`로 저장되어 있습니다.

이를 해결하기 위해서는 `정수형`으로 `형변환`한 후에 계산을 진행해야 합니다.

#### **1-4-2. 코드 작성 (8점)**

- 각 학생별 총점을 계산해주세요.
  - **🔍힌트:** 각 학생의 점수는 가로 방향(=한 행)에 나열되어 있습니다.
- 과목별 최고점을 계산해주세요.
  - **🔍힌트:** 각 과목의 점수는 세로 방향(=각 열)에 나열되어 있습니다.

**💚 Test Case**  

- 출력
```
학생별 총점: [195, 90, 275]
과목별 최고점: [100, 85, 95]
```

````python
### 예시 답안
# 점수 데이터가 담긴 이차원 리스트
scores = [
    ["60", "85", "50"],
    ["20", "40", "30"],
    ["100", "80", "95"]
]

# 1-4-1번의 문제를 해결하여 scores 이차원 리스트를 재정의한 scores_new 생성하기
## 코드로 구현하면 2점, 직접 int로 리스트 만들면 1점
scores_new = [list(map(int, student_scores)) for student_scores in scores]

# 학생별 총점을 구하고 total_scores 리스트에 저장하기
total_scores = [sum(student) for student in scores_new]

print(f"학생별 총점: {total_scores}")

# 과목별 최고점을 구하고 max_scores 리스트에 저장하기
max_scores = []
for i in range(len(scores_new[0])):
  max_scores.append(max(score[i] for score in scores_new))
# max_scores = [max(subject) for subject in zip(*scores_new)]

print(f"과목별 최고점: {max_scores}")
````

## **2-1. 제시문에 따라 문제를 해결해주세요. (6점)**
- 반복문을 이용해 주어진 리스트들의 합을 구하는 함수를 만드세요

- 이때 함수의 이름은 calcsum( )입니다. [3점]

- 위에서 구한 calcsum( )을 이용해 평균을 구하세요.

- 이때 평균에 대한 함수 이름은 calcavg( )입니다. [3점]


-주어진 리스트 : ex_list = [-6, 3, 2, 41, 15, 35, 23, -3]

````python
### 답안

# calcsum( )함수 생성 (3점)
def calcsum(inputlist):
  sum = 0

  for i in inputlist:
    sum = sum + i

  return sum

# calcavg( )함수 생성 (4점)
def calcavg(inputlist):
  avg = None

  sum = calcsum(inputlist)

  if len(inputlist) > 0 :
    avg = sum / len(inputlist)
  else :
    avg = None

  return avg

ex_list = [-6, 3, 2, 41, 15, 35, 23, -3]

# 결과 출력
print("주어진 리스트의 합:\n", calcsum(ex_list))
print("\n주어진 리스트의 평균:\n", calcavg(ex_list))
````

## 2-4. 제시문에 따라 문제에 대한 답을 제시하고 해결해주세요. (10점)

- 아래와 같이 문자열이 하나 주어졌습니다.

> sentence = " apple banana apple orange banana apple"


- 문자열을 공백 기준으로 나눈 후 각 단어의 등장 횟수를 딕셔너리로 반환 (4점)
  - 위 함수의 이름을 count_words(sentence)라고 한다.


- 가장 많이 등장한 단어를 반환하는 함수 만들기(3점)
  - 위 함수의 이름은 most_common(word_dict)이다. 이때, 반환값은 하나이다.

- 특정 단어가 몇 번 등장했는지 반환하는 함수 만들기(3점)
  -위 함수의 이름은 count_word(word_dict, word)이다. 없으면 0 반환.




**💚 Test Case**  
  
[예시]
- 입력
```
word_dict = count_words(sentence)

print(word_dict)

print(most_common(word_dict))

print(count_word(word_dict, "apple"))
```

- 출력
```
{'apple' :3, 'banana':2, 'orange' : 1}
apple
3
```

````python
### 예시 답안

sentence = "apple banana apple orange banana apple"

# split( )을 이용해 공백을 없애고, 각 단어의 등장 횟수를 딕셔너리로 반환하는 함수 생성
def count_words(sentence) :
  words = sentence.split()

  word_dict = {}

  for word in words :
    if word in word_dict :
      word_dict[word] = word_dict[word] + 1
    else :
      word_dict[word] = 1

  return word_dict

# 가장 많이 등장한 단어를 반환하는 함수 생성
def most_common(word_dict) :
  max_count = 0
  max_word = ""

  for word in word_dict :
    if word_dict[word] > max_count :
      max_count = word_dict[word]
      max_word = word

  return max_word

#특정 단어가 몇 번 등장했는지 확인하는 함수 생성
def count_word(word_dict, word) :
  if word in word_dict :
    return word_dict[word]
  else :
    return 0

#실행
word_dict = count_words(sentence)

print(word_dict)

print(most_common(word_dict))

print(count_word(word_dict, "apple"))
````

##문제 3-2 (10점)
-출제 의도


*   자료구조 배열(list) 활용 능력 평가
*   인덱스 기반 탐색 알고리즘 구현
*   반복문 + 조건문을 이용한 Two Sum 문제 해결

정수 배열 nums와 정수 target을 입력받아,
배열에서 두 수의 합이 target이 되는 두 인덱스를 출력하시오.

정답은 순서 상관없이 출력 가능하다.

📌 입력 형식


1.   첫 줄: 배열 요소 (공백으로 구분된 정수)
2.   둘째 줄: target 값

     예)

     2 7 11 15

     9

📌 출력 형식

    [0, 1]  (설명: nums[0] + nums[1] = 9)

💚 Test Case

예시1)

Input: nums = [3,2,4], target = 6

Output: [1,2]

예시2)

Input: nums = [3,3], target = 6

Output: [0,1]

📌 조건



*   같은 인덱스 두 번 사용 금지
*   반드시 하나의 정답만 존재
*   배열 길이: 2 이상
*   for 반복문 사용
*   리스트 인데스 접근 사용

````python
###답안 작성###
nums = list(map(int, input().split()))
target = int(input())

for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] + nums[j] == target:
            print([i, j])
````

##문제 3-3. Counting + Sort 활용 문제 (10점)
-출제 의도


*   자료구조 **리스트(List)** 활용 능력 평가
*   Counting 개념 이해
*   정렬(sort)을 활용한 알고리즘 설계 능력 평가
*   반복문과 조건문을 통한 비교 연산 구현

정수 배열 nums가 주어질 때,
각 원소에 대해 자신보다 작은 수의 개수를 계산하여 리스트로 반환하시오.

조건:


*   j != i
*   nums[j] < nums[i]

📌 함수 형식 (반드시 유지)


```
def smallerNumbersThanCurrent(self, nums):
    """
    :type nums: List[int]
    :rtype: List[int]
    """
```
반환값: 각 위치별 작은 수의 개수를 담은 리스트

💚 Test Case

예시1)

Input: nums = [8,1,2,2,3]

Output: [4,0,1,1,3] (8보다 작은 수: 4개, 1보다 작은 수: 0개, 2보다 작은 수: 1개, 3보다 작은 수: 3개)

예시2)

Input: nums = [6,5,4,8]

Output: [2,1,0,3]

📌 제약 조건


*   2 ≤ nums.length ≤ 500
*   0 ≤ nums[i] ≤ 100|


````python
###답안작성###
def smallerNumbersThanCurrent(self, nums):
        sorted_nums = sorted(nums)
        result = []

        for num in nums:
            count = sorted_nums.index(num)
            result.append(count)

        return result
print(smallerNumbersThanCurrent(nums))
````
